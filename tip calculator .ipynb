```python
#tip calculator
print("Welcome to the tip calculator")
a=float(input("How much is the total bill ? $"))
b=int(input("how much percent of tip would u like to give ? " + "10 , 15 or 12"))
c=int(input("how many people spit the bill? "))
split=(((b/100)*a)+a)/c
formatted_str = f"{split:.3f}"   # used to make the value at n=3,2etc  decimal place.3f means the at what decimal place u want to print
print(f"each person should pay : ${formatted_str}")
      
```
